```python
def odd_v_search(i, j): # 홀수 회문 세로 확인
    global max_length

    length = 0
    cnt = 0

    while True:
        if i - cnt <= -1 or i + cnt >= 100:
            break
        if graph[i - cnt][j] == graph[i + cnt][j]:
            length = 1 + 2 * cnt
        else:
            break
        cnt += 1

    if max_length < length:
        max_length = length

def odd_h_search(i, j): # 홀수 회문 가로 확인
    global max_length

    length = 0
    cnt = 0
    while True:
        if j - cnt <= -1 or j + cnt >= 100:
            break
        if graph[i][j - cnt] == graph[i][j + cnt]:
            length = 1 + 2 * cnt
        else:
            break
        cnt += 1

    if max_length < length:
        max_length = length
        
def even_v_search(i, j): # 짝수 회문 세로 확인
    global max_length

    length = 0
    cnt = 0
    
    if i+1 >= 100:
        return
    
    if graph[i][j] != graph[i+1][j]:
        return
    
    while True:
        if i - cnt <= -1 or i + 1 + cnt >= 100:
            break
        if graph[i - cnt][j] == graph[i + 1 + cnt][j]:
            length = 2 * (cnt+1)
        else:
            break
        cnt += 1

    if max_length < length:
        max_length = length

def even_h_search(i, j): # 짝수 회문 가로 확인
    global max_length

    length = 0
    cnt = 0

    if j + 1 >= 100:
        return

    if graph[i][j] != graph[i][j+1]:
        return

    while True:
        if j - cnt <= -1 or j + 1 + cnt >= 100:
            break
        if graph[i][j - cnt] == graph[i][j + 1 + cnt]:
            length = 2 * (cnt+1)
        else:
            break
        cnt += 1

    if max_length < length:
        max_length = length

for test_case in range(1, 11):
    tc = int(input())
    graph = [list(input()) for _ in range(100)]
    max_length = 0

    for i in range(len(graph)):
        for j in range(len(graph)):
            odd_v_search(i,j)
            odd_h_search(i,j)
            even_v_search(i,j)
            even_h_search(i,j)
    print(f'#{tc}', max_length)
```
